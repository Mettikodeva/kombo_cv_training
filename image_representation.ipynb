```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("zebra.jpg")
h,w,d = img.shape
img = cv.resize(img, (w//5,h//5), interpolation=cv.INTER_NEAREST)
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
```

```python
print("normal image shape: ",img.shape)
print("gray image shape: ",gray.shape)  
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 5))
axes[0].imshow(rgb)
axes[0].set_title("RGB")
axes[0].axis("off")
axes[1].imshow(gray)
axes[1].set_title("GRAY")
axes[1].axis("off")
axes[2].imshow(img)
axes[2].set_title("BGR")
axes[2].axis("off")
axes[3].imshow(hsv)
axes[3].set_title("HSV")
axes[3].axis("off")
plt.show()
```

# Image Thresholding

```python
img_text = cv.imread("text.jpg")
threshval = 150
maxval = 255
img_text = cv.threshold(img_text,threshval,maxval,cv.THRESH_BINARY)
img_text = img_text[1]
```

```python
plt.imshow(img_text, cmap='gray')
```
