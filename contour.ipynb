```python

import cv2 as cv
import numpy as np
import random as rng
from matplotlib import pyplot as plt
```

```python
img = cv.imread("shape.jpg")
h,w,d =img.shape
# img = cv.resize(img, (w//5,h//5))
```

```python
from copy import deepcopy
new_img = deepcopy(img)
h,w,d = img.shape   

img_gray = cv.cvtColor(new_img,cv.COLOR_BGR2GRAY)
```

```python
threshold = 200
# edge detection
canny_image = cv.Canny(img_gray,threshold,threshold*2)

# find countour
contours,hierarchy = cv.findContours(canny_image,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
```

```python
# Draw contours
drawing = np.zeros((canny_image.shape[0], canny_image.shape[1], 3), dtype=np.uint8)
for i in range(len(contours)):
    color = (rng.randint(0,256), rng.randint(0,256), rng.randint(0,256))
    cv.drawContours(drawing, contours, i, color, 2, cv.LINE_8, hierarchy, 0)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 5))
axes[0].imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
axes[0].set_title("img")
axes[0].axis("off")
axes[1].imshow(img_gray, cmap="gray")
axes[1].set_title("gray")
axes[1].axis("off")
axes[2].imshow(canny_image)
axes[2].set_title("canny_image")
axes[2].axis("off")
axes[3].imshow(drawing)
axes[3].set_title("img_drawing")
axes[3].axis("off")
plt.show()
```
