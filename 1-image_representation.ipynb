```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("zebra.jpg")
h,w,d = img.shape
img = cv.resize(img, (w//5,h//5), interpolation=cv.INTER_NEAREST)
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
rgba = cv.cvtColor(img,cv.COLOR_BGR2RGBA)
```

```python
print("normal image shape: ",img.shape)
print("gray image shape: ",gray.shape)  
print("rgba image shape: ",rgba.shape)  
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 5))
axes[0].imshow(rgb)
axes[0].set_title("RGB")
axes[0].axis("off")
axes[1].imshow(gray, cmap="gray")
axes[1].set_title("GRAY")
axes[1].axis("off")
axes[2].imshow(img)
axes[2].set_title("BGR")
axes[2].axis("off")
axes[3].imshow(hsv)
axes[3].set_title("HSV")
axes[3].axis("off")
plt.show()
```

# Image Thresholding

```python
img_text = cv.imread("text.jpg")
plt.imshow(cv.cvtColor(img_text,cv.COLOR_BGR2RGB))
# cv.imshow("img",img_text)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
img_text = cv.imread("text.jpg") #np.uint8
threshval = 150
maxval = 255
# 
# thresholding func
(_,img_bin_inv) = cv.threshold(img_text,threshval,maxval,cv.THRESH_BINARY_INV)
(_,img_bin) = cv.threshold(img_text,threshval,maxval,cv.THRESH_BINARY)
(_,img_trunc) = cv.threshold(img_text,threshval,maxval,cv.THRESH_TRUNC) 
(_,img_tozero) = cv.threshold(img_text,threshval,maxval,cv.THRESH_TOZERO)
(_,img_tozero_inv) = cv.threshold(img_text,threshval,maxval,cv.THRESH_TOZERO_INV)
```

```python
# cv.imshow("img",img_bin_inv)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
fig, ax = plt.subplots(2,3, figsize=(12, 7))
print(ax)
ax[0,0].set_title("original")
ax[0,0].imshow(cv.cvtColor(img_text, cv.COLOR_BGR2RGB))
ax[0,1].set_title("bin")
ax[0,1].imshow(img_bin, cmap="gray")
ax[0,2].set_title("bin_inv")
ax[0,2].imshow(img_bin_inv, cmap="gray")
ax[1,0].set_title("truncate")
ax[1,0].imshow(img_trunc, cmap="gray")
ax[1,1].set_title("tozero")
ax[1,1].imshow(img_tozero, cmap="gray")
ax[1,2].set_title("tozero_inv")
ax[1,2].imshow(img_tozero_inv, cmap="gray")
```
