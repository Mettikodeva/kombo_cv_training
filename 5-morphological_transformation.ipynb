```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread("opening.png")
plt.imshow(img)
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
img_erode = cv.morphologyEx(img,cv.MORPH_ERODE,kernel)
plt.imshow(img_erode)
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
img_dil = cv.morphologyEx(img_erode,cv.MORPH_DILATE,kernel)
plt.imshow(img_dil)
```

## CLOSING

```python
img = cv.imread("closing.png")
plt.imshow(img)
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
img_dil = cv.morphologyEx(img,cv.MORPH_DILATE,kernel)
plt.imshow(img_dil)
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
img_erode = cv.morphologyEx(img_dil,cv.MORPH_ERODE,kernel)
plt.imshow(img_erode)
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
img_gradient = cv.morphologyEx(img_erode,cv.MORPH_GRADIENT,kernel)
plt.imshow(img_gradient)
```
